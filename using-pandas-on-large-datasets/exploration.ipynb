## Technique 2 - Use Efficient Data Types


Numeric data are the easiest to reduce:

| Data Type (bits) | Type       | Maximum Value        | Minimum Value        | Note                 |
|------------------|------------|----------------------|----------------------|----------------------|
| 8-bit            | Integer    | 127                  | -128                 |                      |
| 16-bit           | Integer    | 32,767               | -32,768              |                      |
| 32-bit           | Integer    | 2,147,483,647        | -2,147,483,648       |                      |
| 64-bit           | Integer    | 9,223,372,036,854,775,807 | -9,223,372,036,854,775,808 |               |
| 32-bit           | Floating   | 3.4028235e+38        | -3.4028235e+38       | Precision of ~7 decimal digits |
| 64-bit           | Floating   | 1.7976931348623157e+308 | -1.7976931348623157e+308 | Precision of ~15 decimal digits |

```python
smaller_df.info(memory_usage='deep')
```

```python
# A more efficient schema for our data
schema = {
    "hotel": "category",
    "is_canceled": "boolean",
    "lead_time": "int16",
    "arrival_date_year": "int16",
    "arrival_date_month": "category",
    "arrival_date_day_of_month": "int8",
    "adults": "int8",
    "children": "int8",
    "babies": "int8",
    "customer_type": "category"
}
```

```python
smaller_df.fillna(0, inplace=True)

cast_df = smaller_df.astype(schema).copy()

cast_df.head(10)
```

```python
cast_df.info(memory_usage='deep')
```
